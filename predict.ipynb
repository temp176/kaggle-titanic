```python
import pandas as pd
import csv as csv
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load training data
train_df = pd.read_csv("train.csv", header=0)

# Convert "Sex" to be a dummy variable (female = 0, Male = 1)
train_df["Gender"] = train_df["Sex"].map({"female": 0, "male": 1}).astype(int)
train_df.head(3)

# Complement the missing values of "Age" column with average of "Age"
median_age = train_df["Age"].dropna().median()
if len(train_df.Age[train_df.Age.isnull()]) > 0:
    train_df.loc[(train_df.Age.isnull()), "Age"] = median_age

# remove un-used columns
train_df = train_df.drop(["Name", "Ticket", "Sex", "SibSp", "Parch", "Fare", "Cabin", "Embarked", "PassengerId"], axis=1)
train_df.head(3)
```

```python
# Load test data, Convert "Sex" to be a dummy variable
test_df = pd.read_csv("test.csv", header=0)
test_df["Gender"] = test_df["Sex"].map({"female": 0, "male": 1}).astype(int)

# Complement the missing values of "Age" column with average of "Age"
median_age = test_df["Age"].dropna().median()
if len(test_df.Age[test_df.Age.isnull()]) > 0:
    test_df.loc[(test_df.Age.isnull()), "Age"] = median_age

# Copy test data's "PassengerId" column, and remove un-used columns
ids = test_df["PassengerId"].values
test_df = test_df.drop(["Name", "Ticket", "Sex", "SibSp", "Parch", "Fare", "Cabin", "Embarked", "PassengerId"], axis=1)
test_df.head(3)
```

```python
train_x = train_df.values[:,1:]
train_y = train_df.values[:,0]
```

```python
print("train_x:",train_x.shape,"train_y:",train_y.shape)
```

```python
from tensorflow.python.keras.utils import to_categorical
```

```python
train_y = to_categorical(train_y)
```

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense,Dropout
```

```python
model = Sequential()

model.add(
    Dense(
        units=64, 
        input_shape=(3,),
        activation='relu'
    )
)

model.add(Dropout(0.25))

model.add(
    Dense(
        units=32, 
        input_shape=(3,),
        activation='relu'
    )
)

model.add(
    Dense(
        units=2, 
        activation='softmax'
    )
)
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_adam = model.fit(
    train_x,
    train_y,
    batch_size=32,
    epochs=100,
    validation_split=0.2,)
```

```python
res = model.predict(test_df.values)
```

```python
output = [0 for i in range(len(res))]
for i in range(len(res)):
    if res[i][0] >= 0.5:
        output[i] = 0
    else:
        output[i] = 1
```

```python
submit_file = open("titanic_submit.csv", "w")
file_object = csv.writer(submit_file)
file_object.writerow(["PassengerId", "Survived"])
file_object.writerows(zip(ids, output))
submit_file.close()
```
